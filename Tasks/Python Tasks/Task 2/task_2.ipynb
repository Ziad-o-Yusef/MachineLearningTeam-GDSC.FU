```python
#section2:
#1- Python program to display all numbers within a range except the prime numbers.
l=[]
n=int(input("enter number of numbers you need"))
for i in range(0,n):
    l.append(int(input(f"enter the number {i+1} \t")))

for i in range(0,n) :
    if l[i]==1 or l[i]==2 or l[i]==3:
     continue
    elif l[i]%2==0 or l[i]%3==0:
        print(l[i])  
    else:
        continue
```
