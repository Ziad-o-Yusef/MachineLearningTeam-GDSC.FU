```python
import math as mt

def isPrime(n):
    if(n==0 or n==1):
        return False
    for i in range(2,int(mt.sqrt(n))+1):
        if n % i == 0:
            return False
    return True

for i in range (50):
    if(not isPrime(i)):
        print(i,end=',')
```
