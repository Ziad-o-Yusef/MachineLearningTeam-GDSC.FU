```python
lower_value = int(input (" Enter  Lowest Range Value: "))  
upper_value = int(input (" Enter  Upper Range Value: "))  
  
print ("The Prime Numbers are ")  
for number in range (lower_value, upper_value + 1):  
    if number > 1:  
        for i in range (2, int(number/2)):  
            if (number % i) == 0: 
                print (number) 
            else: 
                break
```
