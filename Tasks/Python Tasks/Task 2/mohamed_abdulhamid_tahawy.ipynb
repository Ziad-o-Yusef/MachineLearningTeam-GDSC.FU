```python
#Section 2 :
#Q1:-
def prime_number (num):
     result=False
     if num > 1:
          for i in range(2, num):
               if (num % i) == 0:
                    result = True
                    break
          return result

start=int(input("enter the start of your range "))
end=int(input("enter the end of your range "))
for i in range(start,end):
     if i==1:
          continue
     if prime_number(i):
          continue
     print(i)
```
