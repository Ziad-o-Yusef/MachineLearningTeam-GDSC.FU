```python
#SECTION 1
#Q1:-
def Grade():
    ur_grade = int(input("Please Enter Ur Garde : "))
    if ur_grade < 25 :
        case = "F"
    elif ur_grade >= 25 :
        case = "Ur Evaluation Is: E"
    elif ur_grade >= 45 :
        case = "Ur Evaluation Is: D"
    elif ur_grade >= 50 :
        case = "Ur Evaluation Is: C"
    elif ur_grade >= 60 :
        case = "Ur Evaluation Is: B"
    elif ur_grade >= 80 :
        case = "Ur Evaluation Is: A"
    print(case )
    return ;
```

```python
Grade()
```

```python
#SECTION 2
#Q1 :-
def nums():
    range_num = int(input("Please Enter Range Num : "))
    for num in range(0,range_num):
        if (num % 2 ) == 0 and num > 2:
            print(num)
    return ;
```

```python
#SECTION 3
#Q1:-
def Sum(*nums):
    new_nums = nums
    sum = 0
    for i in new_nums :
        sum += i 
    print("Summation = " , sum)
```
